```python
# Classification of Education into meaningful categories
credit['EDUCATION'] = pd.Categorical(credit.EDUCATION)
credit["EDUCATION"].cat.categories = ["Others0","Graduate", "University", "High School","Others4","Others5","Others6"]
def recode_education(series): 
    if(series == 'Others0') or (series == 'Others4') or (series == 'Others5') or (series == 'Others6'): 
        return 'Other'
    else:
        return series
credit["EDUCATION"] = credit["EDUCATION"].apply(recode_education) 
```

### Correlation

```python
#building a basic correlation coefficient matrix using pandas
corrMat = credit.corr()
print(corrMat)
```

Credit Default is not likely to occur with history of past payment amount with stronger realtionship with the most recent payments. Rate of default would higher with Bill Amount and Amount of previous payment

### Covariance

```python
#Calculate covariance in Pandas enter the following:
covMat = credit.cov()
print(covMat)
```

Higher the Credit limit, Bill Amount and Amount of previous payment is likely to have reduce the Default. Higher Payment history is likely to reduce credit default rate.

```python
#Converting following columns into Categorical
credit['MARRIAGE'] = pd.Categorical(credit.MARRIAGE)
credit['MARRIAGE'].cat.categories = ["Others", "Married", "Single","Divorced"]
```

```python
# Create a generation bin
generations = [10,20,30,40,50,60,70,80]
# Default vs. Age grouped by Gender
sns.lmplot('AGE', 'default payment next month', data=credit, hue='SEX',x_bins=generations)
```

Both male and female in 70s seems to have higher default rates

```python
# Default vs. Age grouped by EDUCATTION
sns.lmplot('AGE', 'default payment next month', data=credit, hue='EDUCATION',x_bins=generations)
```

People with no education tends to default in 20s where as people who have attended graduate school or high school tends to default in 70s.

```python
# Default vs. Age grouped by Marriage status
sns.lmplot('AGE', 'default payment next month', data=credit, x_bins=generations,hue='MARRIAGE')
```

Rate of default is higher for divorced individual and where as default rate drops with age for single and unknown marriage status

```python
# Create a Credit Limit  bin
limit_range = [0, 25000,50000,100000,150000,250000, 500000, 1000000]
# Default vs. Credit Limit grouped by Gender
sns.lmplot('LIMIT_BAL', 'default payment next month', data=credit, hue='SEX',x_bins=limit_range)
```

Rate of default reduces with higher credit limit for both gender

```python
# Default vs. Credit Limit grouped by EDUCATTION
sns.lmplot('LIMIT_BAL', 'default payment next month', data=credit, hue='EDUCATION', x_bins=limit_range)
```

People with unknown education tends to have lower credit limit

```python
# Default vs. Credit Limit grouped by Marriage status
sns.lmplot('LIMIT_BAL', 'default payment next month', data=credit, x_bins=limit_range,hue='MARRIAGE')
```

Rate of default does not reduce with higher credit limit whereas the default rate falls sharply with higher credit limit
